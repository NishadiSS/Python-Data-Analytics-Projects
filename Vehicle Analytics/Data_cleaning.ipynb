```python
import pandas as pd
```

```python
df=pd.read_csv("auto_with_col.csv")
```

```python
df.head()
```

```python
import numpy as np

#replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
#To check how many number of nan value in the whole data set
df.isnull().sum()
```

```python
#
df.dtypes
```

```python
#normalized-losses  column not object so we make it integer data type

#Calculate the average of the column
avg_norm_loss = df["normalized-losses"].astype("float").mean() 
print("Average of normalized-losses:", avg_norm_loss)

#Replace NaN by mean value
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Replace NaN in "stroke" column by mean.

# calculate the mean vaule for "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
# replace NaN by mean value in "stroke" column
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
#Replace NaN in "bore" column by mean.

# calculate the mean vaule for "bore" column
avg_bore = df["bore"].astype("float").mean(axis = 0)
print("Average of bore:", avg_bore)
# replace NaN by mean value in "bore" column
df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
#Replace NaN in "horsepower" column by mean.

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Replace NaN in "peak-rpm" column by mean.

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#No of doors column have numbers as strings so it convert using frequency method

#To see which values are present in a particular column, we can use the ".value_counts()" method:
df['num-of-doors'].value_counts()
```

```python
#We can see that four doors are the most common type. We can also use the ".idxmax()" method to calculate for us the most common type automatically:
#The replacement procedure is very similar to what we have seen previously

#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#we predit price value
#so if it hase missing value row then we remove all rows 
#Finally, let's drop all rows that do not have price data:

# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
#Convert data types to proper format
df[["bore", "stroke", "peak-rpm","price"]] = df[["bore", "stroke", "peak-rpm","price"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
#save the new csv
df.to_csv('cleaned_data.csv')
```
